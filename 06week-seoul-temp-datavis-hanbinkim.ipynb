```python
# 날짜데이터 추출 (4/8)
# split함수 사용 (최고기온만 그래프로 그리기)
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []

for row in data:
    if row[4] != '':
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            result.append(float(row[4]))
            
f.close()

plt.figure(figsize=(10, 2))
plt.plot(result, 'r')
plt.xlabel('Day (Decembmer)')
plt.ylabel('The highest temperature')
plt.show()
```

```python
# 날짜데이터 추출 (5/8)
# split함수 사용 (최고기온이랑 최저기온 그래프로 그리기)
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))
f.close()

plt.figure(figsize=(10, 2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day (Decembmer)')
plt.ylabel('The highest temperature')
plt.show()
```

```python
# 날짜데이터 추출 (6/8), (7/8)
# split함수 사용 (그래프 다듬기, 글씨체랑 마이너스부호)
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))
f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day (Decembmer)')
plt.ylabel('The highest temperature')
plt.legend()
plt.grid()
plt.show()
```

```python
# 날짜데이터 추출 (8/8)
# split함수 사용 (평균 기온 변화 그래프 그리기)
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

temp_day = []
temp_mean = []
year = []

for row in data:
    #1907년도에는 10월부터 기온 데이터가 기록되어 있어서 평균 기온이 낮음 (분석에서 제외)
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '': # 평균 기온 데이터가 결측치가 아니라면 아래를 수행
            # 1월 1일부터 12월 31일까지의 일별 평균 기온 데이터를 저장
            temp_day.append(float(row[2]))
            if(row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '31'):
                year.append(int(row[0].split('-')[0])) # 연도데이터 저장
                #12월 31일이 되면 지금까지 temp_day에 저장해 둔 평균 기온 데이터의 평균 값을 계산하기
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = [] # 다음 년도의 평균 기온 데이터를 1월 1일부터 새롭게 저장해야 함
                #temp_day 리스트를 비어 있는 리스트로 초기화
f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family = 'Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year, temp_mean, color='k', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```
