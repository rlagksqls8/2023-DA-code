```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[10,20,30,40], color='skyblue', label='skyblue')
plt.plot([1,2,3,4],[40,30,20,10], color='red', label='red')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[10,20,30,40], color='skyblue', linestyle='--', label='dashed')
plt.plot([1,2,3,4],[40,30,20,10], color='red', ls=':', label='dotted')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[10,20,30,40], color='skyblue', linestyle='--', marker='d', label='diamond')
plt.plot([1,2,3,4],[40,30,20,10], color='red', ls=':', marker='^', label='triangle up')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[10,20,30,40], color='skyblue', linestyle='--', marker='d', linewidth=3, label='thick line')
plt.plot([1,2,3,4],[40,30,20,10], color='red', ls=':', marker='^', lw=0.5, label='thin line')

plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```

```python
import numpy as np

x = np.arange(-10,10,0.1) # -10 부터 10까지 0.1씩 증가
plt.plot(x, x**2)

plt.xticks(np.arange(-10,11,2)) #X눈금
plt.yticks(np.arange(0,101,20)) #y눈금
plt.grid(True)

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('선 그래프 예제')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200) #0에서 2*pi 까지 경계선 포함 200등분
y = np.sin(x)
plt.plot(x, y)

plt.grid(True)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('삼각 함수')
```
